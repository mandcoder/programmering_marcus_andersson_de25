```python
from rectangle import Rectangle
from circle import Circle
from shape2dplotter import Shape2dPlotter

circ1 = Circle(x=5, y=5, radius=2)

recta1 = Rectangle(x=9.6, y=7, width=5, height=2)

plotter = Shape2dPlotter()

plotter.add_shapes(circ1)

plotter.add_shapes(recta1)

plotter.draw_all()
```
