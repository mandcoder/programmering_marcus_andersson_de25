## üß© Steg 1: L√§s in filen och bygg dictionaryn 
Du ska l√§sa in textfilen, dela varje rad vid kommatecknet och bygga upp en dictionary.
-  L√§gg till felhantering om filen inte finns (`try` / `except FileNotFoundError`)



````python
import pandas as pd

pd.__version__
````

````python

# üëâ Write your code here

# lets create a dictionary

books = {}

# lets read in the file.

with open("books.txt", "r") as booksfile:

    for line in booksfile:

        line = line.strip().split(",")

        # lets create the dictionary

        books[line[0]] = {"author": line[1], "year": line[2]}


````


## ‚úÖ Steg 2: Testa resultatet
Skriv ut dictionaryn och kontrollera att den har r√§tt struktur.


````python
# üëâ Write your code here

books
````

````python
df = pd.DataFrame(books).T # = transpose
df["year"] = df["year"].astype(int)
df.reset_index(inplace=True)
df.rename(columns={"index": "title", "author": "author", "year": "year" }, inplace=True)

"""man m√•ste grupper n√§r man g√∂r flera df √§ndringar"""

(

df.style.hide(axis="index")
.set_properties(**{'text-align': 'left'})
.set_table_styles(
    # selector --> vilket objekt man vill styla. "th" √§r table header
    # props --> lista med CSS-egenskaper (name, value)-tupler
    [{"selector": 'th', "props": [('text-align',"left")]}]
)

.set_caption("üìö Book Collection")

)
````
