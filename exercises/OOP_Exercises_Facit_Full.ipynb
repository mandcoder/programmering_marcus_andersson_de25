## üß† Uppgift 1 ‚Äì Skapa en grundklass

3. Spara `name` direkt, men g√∂r `age` till ett skyddat attribut (`_age`).

```python
class Person:

    def __init__(self, name, age):
        self.name = name
        self._age = age
```

## üß† Uppgift 2 ‚Äì L√§gg till getter och setter

1. √Ñndra klassen **Person** s√• att den anv√§nder `@property` och `@<property>.setter` p√• ett korrekt s√§tt.  
2. Skapa en **property** f√∂r `age`.  
3. Getter-metoden ska returnera `_age` och skriva ut `"Getter called"`.  
4. Setter-metoden ska skriva ut `"Setter called"` och kontrollera att v√§rdet √§r mellan `0` och `124`.  
   - Om v√§rdet inte √§r giltigt, kasta ett `ValueError` med meddelandet `'Age must be between 0 and 124'`.

```python
class Person:

    def __init__(self, name, age):
        self.name = name
        self.age = age   # aktiverar settern vid skapandet

    @property
    def age(self):
        print("Getter called")
        return self._age

    @age.setter
    def age(self, value):
        print("Setter called")
        if not (0 <= value <= 124):
            raise ValueError("Age must be between 0 and 124")
        self._age = value

```

## üß† Uppgift 3 ‚Äì L√§gg till __repr__ och testa klassen
1. Skapa en `__repr__`-metod som visar namn och √•lder p√• ett tydligt s√§tt.
2. Skapa ett objekt `p1` med ett namn och en giltig √•lder.
3. Testa att √§ndra √•lder till ett nytt v√§rde.
4. Testa att ge ett ogiltigt v√§rde (t.ex. `-10`) och se att felet f√•ngas.

```python
class Person:

    def __init__(self, name, age):
        self.name = name
        self.age = age   # aktiverar settern vid skapandet

    @property
    def age(self):
        print("Getter called")
        return self._age

    @age.setter
    def age(self, value):
        print("Setter called")
        if not (0 <= value <= 124):
            raise ValueError("Age must be between 0 and 124")
        self._age = value

    def __repr__(self):
        # Visar objektet p√• ett tydligt och utvecklarv√§nligt s√§tt
        return f"Person(name='{self.name}', age={self._age})"


# --- TESTA KLASSEN ---

# 1. Skapa objekt
p1 = Person("Kalle", 25)

# 2. Visa objektet
print(p1)

# 3. √Ñndra √•lder till ett nytt giltigt v√§rde
p1.age = 40
print(p1)

# 4. Testa ett ogiltigt v√§rde
try:
    p1.age = -10
except ValueError as e:
    print("Fel f√•ngat:", e)
```

## 2. Car class med metod
H√§r √∂var du p√• att skapa en metod (`drive`) som anv√§nder klassens attribut.

```python
class Car:
    def __init__(self, model, year, color, is_electric):
        self.model = model
        self.year = year
        self.color = color
        self.is_electric = is_electric

    def drive(self):
        print(f'The {self.color} {self.model} is now driving.')

    def __repr__(self):
        return f"Car(model='{self.model}', year={self.year}, color='{self.color}', is_electric={self.is_electric})"

car1 = Car('Audi', 2020, 'Black', False)
car2 = Car('Tesla Model 3', 2023, 'White', True)

car1.drive()
car2.drive()
print(car1)
```

## 3. InsurancePolicy class
H√§r l√§r du dig skillnaden mellan `__str__` och `__repr__`.
`__str__` anv√§nds f√∂r anv√§ndarv√§nlig text, medan `__repr__` anv√§nds f√∂r utvecklare/debugging.

```python
class InsurancePolicy:
    def __init__(self, policy_number, customer_name, insurance_type, premium_amount, status):
        self.policy_number = policy_number
        self.customer_name = customer_name
        self.insurance_type = insurance_type
        self.premium_amount = premium_amount
        self.status = status

    def __repr__(self):
        return (
            f"InsurancePolicy(policy_number='{self.policy_number}', "
            f"customer_name='{self.customer_name}', "
            f"insurance_type='{self.insurance_type}', "
            f"premium_amount={self.premium_amount}, "
            f"status='{self.status}')"
        )

    def __str__(self):
        return (
            f"{self.policy_number}: {self.insurance_type} insurance for "
            f"{self.customer_name} - {self.premium_amount} SEK ({self.status})"
        )

policy1 = InsurancePolicy('LF-3456', 'Anna Svensson', 'Home', 4500, 'Active')
policy2 = InsurancePolicy('LF-0987', 'Sara Karlsson', 'Car', 3000, 'Pending')

print(policy1)
print(repr(policy2))
```

## 4. Student class
H√§r skapar du en enkel klass med attribut och metod.
Metoden `study()` anv√§nder klassens egna data f√∂r att skriva ut text.

```python
class Student:
    def __init__(self, name, type_of_study, passed, school, age):
        self.name = name
        self.type_of_study = type_of_study
        self.passed = passed
        self.school = school
        self.age = age

    def study(self):
        print(f'{self.name} studies {self.type_of_study} at {self.school}.')

    def __repr__(self):
        return (
            f"Student(name='{self.name}', "
            f"type_of_study='{self.type_of_study}', "
            f"passed={self.passed}, "
            f"school='{self.school}', "
            f"age={self.age})"
        )

s1 = Student('Anna', 'Data Engineer', True, 'STI', 23)
s2 = Student('Breda', 'UX', True, 'STI', 33)

s1.study()
print(s2)
```
