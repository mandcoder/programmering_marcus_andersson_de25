### Co-locate logic with classes and functions

```python
# lets creata a function to create an age_bucket for our customer data

customer_data = [
    {"name": "customer_1", "id": 1, "age": 100 },
    {"name": "customer_2", "id": 2, "age": 42},
    {"name": "customer_5", "id": 5, "age": 25},
    {"name": "customer_6", "id": 6, "age": 19}
]

def get_age_bucket(customer):
    
    customer_age = customer['age']
    if customer_age > 85:
        return '85+'
    elif customer_age > 50:
        return "50-85"
    elif customer_age > 30:
        return "30-50"
    else:
        return "0-30"
    
for customer in customer_data:
    print(customer['age'], get_age_bucket(customer))
    
```

```python
class Pipeline:
    
    def __init__(self, pipeline_type):
        
        self.pipeline_typ = pipeline_type
        
    def extract(self):
        print(f"Data is being extracted for {self.pipeline_typ}")
        
    def transform(self):
        
        print(f"Data is being transformed for {self.pipeline_typ}")
    
    def load(self):
        print(f"Data is being loaded for {self.pipeline_typ}")
    
    def run(self):
        self.extract()
        self.transform()
        self.load()

# we create an object of type Pipline class

p1 = Pipeline("customer_pipeline")

p2 = Pipeline("customer_pipeline")
    
p1.run() 
print("#"*50)
p2.run()
```

***Customer Order Analysis:***  
Write python code that processes a list of customer orders to calculate the total  
revenue and find the top 3 most frequent customers.  

We have a list of orders, where each order is a dictionary with keys:  
`customer_id`, `product`, `quantity`, and `price`.  

**Formula:**  
`revenue = quantity * price`  

The frequency of a customer is defined as the number of orders.

```python
orders_list = [
    {"customer_id": "C001", "product": "laptop", "quantity": 2, "price": 1200.00},
    {"customer_id": "C002", "product": "mouse", "quantity": 1, "price": 25.99},
    {"customer_id": "C001", "product": "keyboard", "quantity": 1, "price": 89.50},
    {"customer_id": "C003", "product": "monitor", "quantity": 1, "price": 299.99},
    {"customer_id": "C002", "product": "laptop", "quantity": 1, "price": 1200.00},
    {"customer_id": "C004", "product": "headphones", "quantity": 3, "price": 79.99},
    {"customer_id": "C001", "product": "webcam", "quantity": 1, "price": 45.00},
    {"customer_id": "C003", "product": "mouse", "quantity": 2, "price": 25.99},
    {"customer_id": "C002", "product": "speaker", "quantity": 1, "price": 150.00},
    {"customer_id": "C005", "product": "tablet", "quantity": 1, "price": 399.99}
]
```

### Calculate total revenue

```python
custom_counter = {}

for order in orders_list:
    cust = order["customer_id"]
    if cust in custom_counter:
        custom_counter[cust] += 1
    else:
        custom_counter[cust] = 1

sorted_list = sorted(custom_counter.items(),
                     key=lambda item: item[1],
                     reverse=True)[:3]

top_3_customers = dict(sorted_list)

print("\nTop 3 customers by order count")
print("-"*30)
for key, value in top_3_customers.items():
    print(f"{key} | {value} orders" )
```

```python
from pprint import pprint

email_list = [
    "john.doe@company.com",
    "jane.smith@email.co.uk",
    "invalid-email",
    "bob@gmail.com",
    "alice.brown@company.com",
    "john.doe@company.com",
    "missing@domain",
    "test@example.org",
    "@nodomain.com",
    "jane.smith@email.co.uk",
    "valid.user@site.net",
    "no-at-symbol.com",
    "another@test.io"
]

def email_checker(email_list):
    valid = []
    notvalid = []

    for email in email_list:
        if "@" in email:
            pos_at = email.find("@")
            pos_dot = email.rfind(".")   # hittar sista punkten

            # 1️⃣ inget före @
            if pos_at == 0:
                notvalid.append(email)

            # 2️⃣ finns ingen punkt efter @
            elif "." not in email[pos_at:]:
                notvalid.append(email)

            # 3️⃣ annars giltig
            else:
                valid.append(email)
        else:
            notvalid.append(email)

    validated_emails = {
        "valid_emails": valid,
        "invalid_emails": notvalid
    }
    return validated_emails


validated_emails = email_checker(email_list)
pprint(validated_emails)
```
