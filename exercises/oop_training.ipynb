##### üîπ exempel i verkligheten

```python
class Car:
    def __init__(self, brand, color):

        #these are attributes
        self.brand = brand
        self.color = color
        pass

    def drive(self):
        # this is a method ( a function that belongs to the class)
        print(f"{self.brand} is driving...")

# create (instantiate) objects from the class
car1 = Car("volvo", "blue")
car2 = Car("Audi", "red")

# use their methods 

car1.drive()
car2.drive()
```

üß© Vad √§r \_\_repr\_\_

**repr** √§r en specialmetod (eller ‚Äúdunder-metod‚Äù, eftersom den har dubbla underscore \_\_ f√∂re och efter).<br>
Den anv√§nds f√∂r att beskriva hur objektet ska visas n√§r du skriver ut det i terminalen, eller ser det i t.ex. en lista eller DataFrame


Utan \_\_repr\_\_

üîπ Med \_\_repr\_\_


```python
class Student:
    def __init__(self, name, school, grade):
        
        # these are atributes
        self.name = name
        self.school = school
        self.grade = grade
    
    def __repr__(self):
        return f"Student(name='{self.name}', school='{self.school}', grade='{self.grade}')"
    
student1 = Student("Anna", "STI", "C")
print(student1)

        
```

### üîπ Skillnaden mellan `__repr__` och `__str__`

| üéØ **Syfte**               | F√∂r **utvecklare** (debugging/loggning) | F√∂r **anv√§ndare** (presentation/utskrift) |

| üí° **Om `__str__` saknas** | Python anv√§nder `__repr__` ist√§llet     | Ingen skillnad om bara `__repr__` finns   |
