```python
class LotteryTicket:
    def __init__(self, ticket_id, game_type, owner, amount_won):
        # Instance attributes
        self.ticket_id = ticket_id
        self.game_type = game_type
        self.owner = owner
        self.amount_won = amount_won   # aktiverar settern!

    @property
    def amount_won(self):
        return self.__amount_won

    @amount_won.setter
    def amount_won(self, value):
        if value < 0:
            print("Error: Amount cannot be negative.")
            self.__amount_won = value
        else:
            self.__amount_won = value

    def check_win(self):
        if self.amount_won > 0:
            return f"Grattis! Du vann {self.amount_won} SEK!"
        else:
            return "TyvÃ¤rr ingen vinst denna gÃ¥ng."

    def __repr__(self):
        return (f"LotteryTicket(ticket_id='{self.ticket_id}', "
                f"game_type='{self.game_type}', "
                f"owner='{self.owner}', amount_won={self.amount_won})")

    def __str__(self):
        return (f"Biljett {self.ticket_id} â€“ {self.game_type} fÃ¶r "
                f"{self.owner} (vinst: {self.amount_won} SEK)")
# --- Test the class ---
t1 = LotteryTicket("SS-2025-001", "Lotto", "Anna", 0)
t2 = LotteryTicket("SS-2025-002", "Triss", "Erik", 100000)
t3 = LotteryTicket("SS-2025-003", "Keno", "Sara", -500)

print(t1)
print(t2.check_win())
print(repr(t2))
t3.amount_won = 5000
print(t3.amount_won)
```
