```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,10,100)

y = 2*x+5 + np.random.normal(0,2 ,size=len(x))
class AnnotatePlot:
    np.random.seed(42)

    print(f"{x[:5] = }")
    print(f"{x[-5:]= }")
    print(f"{x.shape = }")

    fig, ax = plt.subplots(1)
    ax.plot(x,y)
    ax.set(title = "cool stock")
    ax.annotate("big dipp here", xytext=(9,10), xy=(7.5,14.2), arrowprops={"arrowstyle": "->"})
    
```

```python
class PlotAnnotator:
    def __init__(self, fig, ax):
        self.fig = fig
        self.ax = ax

    def arrow(self):
        self.ax.annotate(
            "big dipp here",
            xytext=(9, 10),
            xy=(7.5, 14.2),
            arrowprops={"arrowstyle": "->"},
        )
fig, ax = plt.subplots(1)

ax.plot(x,y)

annotator = PlotAnnotator(fig, ax)
```
