```python
from collections import Counter

placeholder_header_name = ""
dna_rader = {}  # detta blir main container för att alla sequencer

with open("dna_raw.txt", "r") as dna_line:

    for line in dna_line:

        line = line.strip()  # Ta bort \n

        if line.startswith(">"):

            placeholder_header_name = line

            # Initiera en tom sträng om det är första gången rubriken dyker upp
            if placeholder_header_name not in dna_rader:
                dna_rader[placeholder_header_name] = ""
        else:
            # Lägg till raden i den aktuella rubrikens text (lägg ihop flera rader)
            dna_rader[placeholder_header_name] += line

# Nu räknar vi frekvenser
for rubrik, dna_letter in dna_rader.items():
    frekvenser = Counter(dna_letter.lower())
    dna_rader[rubrik] = dict(frekvenser)

dna_rader
```

```python
# slimmad version

dna_dictionary = (
    {}
)  # thisd will be the main container holding all seqeunces letter counts

sequence_id = ""  # placeholder to keep track on current sequence id, like ">seq1"

# read in the file
with open("dna_raw.txt", "r") as dnafile:

    for line in dnafile:

        line = line.strip()  # Remove leading/trailing spaces and newlines "\n"

        if line.startswith(
            ">"
        ):  # check if the line starts with ">" which symbolice a new sequence

            sequence_id = line  # save sequence id

        # if its not a seqeunce id , than we count the letter
        else:
            # use setdefault() to initialize a counter if the key not exist yet
            # than update() it with counts from the current DNA line
            dna_dictionary.setdefault(sequence_id, Counter()).update(line.lower())
    # after the file is fully read, convert all counter object to real dictionarys
    dna_dictionary = {seq: dict(counts) for seq, counts in dna_dictionary.items()}

# display the final dictionary containing all seqeunces and their letter frequencies
dna_dictionary
```
