## Räkna ut den totala föräljningen för laptops

#### Börja med att lista alla ordrar för laptops

```python
# visa alla rader med produkt typ laptop och spara dom i en ny dataframe med metoden .copy()
# detta medföra att laptops_df blir en egen datafram inte en vy.

laptops_df = df[df["product"] == "laptop"].copy()

laptops_df
```

#### Räkna den totala försäljningen av laptops

```python

# börjar med att lägga till en ny kolumn med total per order

laptops_df["total"] = laptops_df["price"] * laptops_df["quantity"]

total_sale_laptop = laptops_df["total"].sum()

print(f"Total laptops sale ${total_sale_laptop:,.2f}")
```
