**list comprehension<br>**

A concise way to create lists in Python<br>
Compact and easier to read than traditional loops<br>
*Syntax: [expression for value in iterable if condition]*

```python
# list comprehension = A concise way to create lists in Python
#                      Compact and easier to read than traditional loops
#                      [expression for value in iterable if condition] if conditions is optional

import random as rnd

#gen_number = rnd.randint(1,11)

list_gen_number = [rnd.randint(1,6) for z in range(1,11)]

list_gen_number



```

```python
m_numbers = [-1,2,5,6,7,-6]

pos_numbers = [num for num in m_numbers if num >= 0]

pos_numbers


```

```python
grades = [19,34,50,60,23,10,20,32]

passed = [grade for grade in grades if grade >=30]

passed
```

```python
# dictionaries comprehension = A concise way to create lists in Python
#                      Compact and easier to read than traditional loops
#                      [expression for value in iterable if condition] if conditions is optional


cities_in_F = {'New York': 32,'Boston': 75,'Los Angeles': 100, 'Chicago': 50}

cities_in_C = {key: round(((value-32)*(5/9))) for (key, value) in cities_in_F.items()}

print(cities_in_C)
```

```python
weather = {"New York": "snowing", "Boston": "sunny", "Los Angeles": "sunny", "Chicago": "cloudy"}
# expression for value in iterable if condition] if conditions is optional
sunny_weather = {key: value for (key,value) in weather.items() if value == "sunny"}

print(sunny_weather)



```

```python
cities_in_F = {'New York': 32,'Boston': 75,'Los Angeles': 100, 'Chicago': 50}

desc_cities = {key: ("Warm" if value >= 40 else "COLD") for (key, value) in cities_in_F.items()}

print(desc_cities)
```
