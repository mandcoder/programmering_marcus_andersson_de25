# Lecture 5b - while loop
- loop through the same block of code for several times
- When number of looping is unknown

```python
# a simple example

n = 0

while n < 10: # repeat multiple until the condition is not met.
    print(n)
    n += 2 # detta är samma som n = n + 1
```

#### Example of oil leakage
- in year 0, for beginning, we have 80000 birds
- due to oil leakage, 50% of birds will die every year ( note that the nr of dead birds is not an absolute number)
- when will the number of birds become 8000?
- in this example, we dont know how many years need to be repeated to reach  this number (8000 birds)

```python
# initial values for the yaer and birds variables

birds = 80000

year = 0



while birds > 8000: # provide a condition , not the number of looping times with the while statement
    print(f"Currently, this is year {year} and the number of the birds is {birds}")
    birds /= 2 # use the assignment operator /=, to simplify the code: birds = birds/2
    print(f"At the end of the year {year} and the number of birds is {birds}")
    print("------------------------------------------------------------------")
    year += 1

print(f"It takes {year} years for the number of birds to drops to 8000")
```

#### Infinite while loop and break statement
- the special thing about an infinite while loop is always True
- loop through the same block of codes for several times
- number of looping is not known when starting a while loop
- instead of having the condition at the begining of the while loop, the condition is still here, but it is evaluated within the block of the codes we want to repeat.
- the motivation/use case: wheter the condition is met will be decided during iteration.

```python
# turning the simple example above to an infinite whil loop

i = 0 
while True:
    print(i)
    i +=1

    if i >= 5: # detta medför att den inte kommer att skriva ut 5.  
        break 
```
