### Clean the quotes file

```python
# wi will use regular expression to clean string contents
import re
```

```python
# create an empty list for saving quotes

quotes = []

# initiate a quote number

i = 1 
```

```python
# open the file
# comparing the use of absolute and relative paths

# print out all the content at once

with open("quotes.txt", "r") as f_read:
    quotes = f_read.read()


print(f_read) # what you can see information about the file object
print(quotes) # what you can see from applying  the read() method on the same file


```

```python
# loop through lines in the file contents
with open("quotes.txt", "r") as f_read:
    for quote in f_read:
        
        # data cleaning

        quote = quote.strip(" \n") # replace the quote with a modified quote
        quote = re.sub(" +" ," ",quote) # replace multiple spaces with one space


        # print each line
        print(quote)
```

```python
# open two files  (one with r-mode and another  with w mode)
i = 1
with open("quotes.txt", "r") as f_read, open("quotes_clean.txt", "w") as f_write:
    for quote in f_read:
        
        # data cleaning

        quote = quote.strip(" \n") # replace the quote with a modified quote
        quote = re.sub(" +" ," ",quote) # replace multiple spaces with one space

        # print each line
        #print(quote)

        if quote != "":
            f_write.write(f"{i}. {quote}\n")
            i += 1

            
```
