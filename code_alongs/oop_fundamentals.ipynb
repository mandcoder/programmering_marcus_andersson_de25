## OOP_fundamentals

- object oriented programming (oop)


```python
class admission:
    # dunder init __ is initializer
    # this method will be run when you instantiate a instance from this class
    
    def __init__(self, school, name, program = "cool program"):
        print("dunder init is runing")
        
        # set the attributes to values from the arguments
        self.school = school
        self.name = name
        self.program = program
        
# instantiate the instance person1 that is a type of admission 
person1 = admission("STI", "Ceda")

person1.school
```

```python
class Car:

    def __init__(self, brand, colour, doors):

        self.brand = brand
        self.colour = colour
        self.doors = doors


car1 = Car("Audi","white", 4)

car1.brand
```

```python
class Student:

    def __init__(self, name, type_of_study, pass1, school, age):
        # set the attributes to values from the arguments
        self.name = name
        self.type_of_study = type_of_study
        self.pass1 = pass1
        self.school = school
        self.age = age

    # method
    def study(self):
        print(f"{self.name} studies {self.type_of_study}")

    
    def study(self):
        print(f"{}")

    def __repr__(self):
        # Example: Student(name='bobbo', type_of_study='economics', school='Cool School', age=31, pass1=False)
        return f"Student(name='{self.name}', type_of_study='{self.type_of_study}', school='{self.school}', age={self.age}, pass1={self.pass1})"


# create objects
s1 = Student(name="anna", type_of_study="data_engineer", pass1=True, school="STI", age=23)
s2 = Student(name="breda", type_of_study="UX", pass1=True, school="STI", age=33)

s2

```
