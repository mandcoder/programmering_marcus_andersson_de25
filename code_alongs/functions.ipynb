#### positional arguments

```python
def my_fruits(fruit_like, fruit_hate):
    print(f"I like to eat {fruit_like}")
    print(f"But i hate to eat {fruit_hate}")
    return None # python will add this line if you dont have return block


```

```python
# In this case i like to eat orange
# check out what is the problem in the order of the arguments below
my_fruits("orange", "apple")
```

#### Keyword arguments

```python
# if you dont wnat to depand  on the position of arguments


my_fruits(fruit_hate="apple", fruit_like="orange")
```

```python
# due to the error, we need to think of other ways to take care of unknown nr arguments.
my_fruits("apple", "orange", "strawberry")

```

#### Unknown number of positional arguments

```python
# suppose that i want to print out all fruits that are inputed as arguments to a function

# we will use a simple * in the parameter

def print_fruits(*fruits): # * has two meanings: positional arguments, the nr of arguments will be unknown
    for fruit in fruits:
        print(fruit)
    return type(fruits)
        


    
```

```python
print_fruits("apple","orange","kiwi")
```

#### unknown number of keyword arguments

```python
def print_nr_fruits(**fruits):
    for key, value in fruits.items():
        print(key, value)
    return type(fruits)



```

```python
print_nr_fruits(fruit1="orange",fruit2="apple") # key is provided when calling the function
```
