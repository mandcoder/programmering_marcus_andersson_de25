### Lecture 6c - Set
- object for those working with mathematical sets

Create a set

```python
# create a list of emails with duplicates

emails = ["anna@example.com", "bobo@example.com","bobo@example.com", "carla@example.com"]

emails
```

```python
# convert emails to a set
# its return only unique value

emails = set(emails)
emails
```

Insersect of two sets

```python
# create two sets

first_month_customers = {"alice", "bob", "carla", "daniel"}
second_mont_custmoers = {"carla", "daniel", "elena"}


```

```python
returning_customers = first_month_customers.intersection((second_mont_custmoers))

returning_customers
```

Union of two sets

```python
all_customers = first_month_customers.union(second_mont_custmoers)

all_customers
```

Except

```python
onetime_customer = all_customers - returning_customers

onetime_customer
```
