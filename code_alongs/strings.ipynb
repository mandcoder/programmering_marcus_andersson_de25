## concatenate strings

```python
# concatenate multiple strings with + operator
# remember + operator has another meaning when it comes  to numeric variables

firstname = "clark"
lastname = "nickelsson"

#concatenate witout space

fullname = firstname + lastname
print(fullname)
```

```python
# concatenate also with
space = " "
fullname_with_space = firstname + " " + lastname # we can replace the space variable just with " "
print(fullname_with_space)
```

```python
# multiple f-string for concatenating many strings

work = "Datalyses AB"

adress = "Down square 120c"

# f-strings with one line
print(f"""{fullname_with_space} is working in a company called {work}.
And his adress is {adress}.""")
```

```python
# contatenate with loop

fruits_today = "apple" + "," + "orange" + "," + "strawberry"

print(f"Todays fruits: {fruits_today}")

fruit_today_list = ["apple,", "orange", "strawberry"]
for fruit in fruit_today_list:
    print(fruit)
```

## Indexing

## indexing  & slicing
- an operation to access "an item"
- what is an item in the context of strings? it is one of the character inside the string
- we can use [] 
- slicing is also possible if we want to access multiple character
- we can use ":" operator for slicing




```python
# index
print(fullname_with_space)
print(f"the first letter of the string is: {fullname_with_space[0]}")
print(f"the last letter is {fullname_with_space[9]}")
```

```python
# slicing

print(f"The first five letters: {fullname_with_space[0:5]}")

# slicing reversely, not used so often
print(f"The last three letters: {fullname_with_space[-1:-6:-1]}")


```

### Some common examples to clean strings
- make use of string methods
- a good introduction to many common string methods:

- syntax for using function on string variable
    - len(string_variable)
- syntax of using method on string variable
    - string_variable.upper()


```python
# splitting a string

subjects = "Math, biology, CheMistry. "

subjects = subjects.split(",") # this will be a list when you split, subjects go from string to list
subjects
```

```python
# strip off leading and trailing space
# check out also rstrip() and lstrip()

for subject in subjects:
    print(subject.strip()) # .strip() takes away spaces

```

```python
# making upper case letter

for subject in subjects:
    print(subject.strip().upper())
```

## Regular expression
- look for a pattern
- make use of different functions in "re" package
- syntax of using function from an imported package
    - re.findall() --> imported_package.function
- check out W3school re for more information

```python
import re
```

```python
text = "det var en gång en katt som på sig en hatt."

list_word_att = re.findall("att",text) # type what you want to find and from where

list_word_att
```

```python
personal_number = "Ida: 19990102-2345, Berit: 890909-9875"

personal_number_eight = re.findall(r"\d{6}-\d{4}", personal_number) # r is raw string

personal_number_eight
```
